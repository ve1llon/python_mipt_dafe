# Домашняя работа №1. N-мерный вектор. 

В данном задании вам предстоит реализовать класс Vector, который предоставляет пользователю функционал вектора n-мерного пространства.

### Критерии оценивания:

- Оценивается оформление кода; Старайтесь писать аккуратно и давать осмысленные имена переменным; Оценка за оформление будет выставляться с помощью PyLint;
- Оценивается правильность реализованных функций; Правильность реализованных функций будет оцениваться с помощью заготовленных юнит-тестов; Задание не принимается, если хотя бы один из тестов не пройден; Изменять тесты ЗАПРЕЩЕНО; 

## Описание задания:

Вы должны реализовать следующие функции, выполнив ряд требований к ним:

- **\_\_init\_\_**:   
    - вектор должен конструироваться из любого итерируемого объекта; Если на вход данной функции подается объект, отличный от итерируемого, вы должны бросить исклечение `TypeError` со следующим сообщением: *type is not iterable; iterable object was expected;*, где type - это тип объекта, поданного на вход (если на вход подан int, то type - это int, если str, то str и т.д.);
    - для хранения данных мы используем [array](https://docs.python.org/3/library/array.html), хранящий числа с плавающей точкой, т.е. вектор не может быть сконструирован из итерируемого объекта, содержащего произвольные значения; В случае, если на вход будет передан итерируемый объект с произвольными значениями, отличными от чисел, вы должны бросить исключение `TypeError` со следующим сообщением: *unsupported elements type; real numbers were expected;*
- **\_\_repr\_\_**:  
    - функция должна возвращать строковое представление вектора, удобное для отладки формата: *Vector([1, 2, 3, 4, 5, ...])*; Если вектор содержит не больше пяти компонент, то сообщение должно иметь следующий вид: *Vector([1, 2, 3, 4, 5])*, иначе используется вариант с многоточием;
- **\_\_str\_\_**: 
    - функция должна возвращать строковое представление объекта, удобное для печати: *(1, 2, 3, 4, 5, 6)*; Т.е. строка должна содержать все компоненты вектора, окруженные круглыми скобками;
- **\_\_len\_\_**: 
    - функция должна возвращать количество элементов в векторе;
- **\_\_abs\_\_**: 
    - функция должна возвращать L2-норму вектора (длину вектора);
- **\_\_bool\_\_**: 
    - функция должна возвращать логическое представление вектора: True - если длина вектора ненулевая, иначе - False;
- **\_\_getitem\_\_**: 
    - перегрузка оператора `[]`, для получения элемента по индексу;
    - в качестве индексов должна поддерживать только объекты типа Integral и slice; если тип переданного индекса отличен от перечисленных, функция должна бросать исключение TypeError с сообщением: *unsupport index type: type; Vector indices must be integers or slices;*, где type - это тип индекса, поданного на вход функции;
    - если тип индекса - Integral, функция должна вернуть элемент, лежащий по данному индексу; в этом случае возвращаемое значение - число;
    - если тип индекса - slice, функция должна вернуть новый объект класса Vector, компоненты которого - компоненты, полученные с помощью среза; в этом случае возвращаемое значение - Vector;
    - если индекс выходит за границы Vector'a, функция должна бросить исключение IndexError с сообщением: *Vector index out of range;*;
- **\_\_setitem\_\_**:  
    - перегрузка оператора `[]`, для установки нового элемента по индексу;
    - в качестве индексов должна поддерживать только объекты типа Integral и slice; если тип переданного индекса отличен от перечисленных, функция должна бросать исключение TypeError с сообщением: *unsupport index type: type; Vector indices must be integers or slices;*, где type - это тип индекса, поданного на вход функции;
    - если тип индекса - Integral, тип переданного значения должно быть Real; иначе бросить исключение TypeError со следующим сообщением: *unexpected value type: type; real number was expected;*, где type - это тип объекта, поданного на вход;
    - если тип индекса - slice, тип переданного значения должно быть Vector; иначе бросить исключение TypeError со следующим сообщением: *unexpected value type: type; Vector was expected;*, где type - это тип объекта, поданного на вход;
    - если индекс выходит за границы Vector'a, функция должна бросить исключение IndexError с сообщением: *Vector index out of range;*;
- **\_\_eq\_\_**: 
    - перегрузка оператора `==`;
    - если тип принимаемого значения отличен от Vector, функция возвращает значение NotImplemented (подумайте, почему?);
    - если переданный объект типа Vector имеет одинаковую длину с нашим вектором, а также покомпонентно равен ему, то функция должна вернуть значение True, иначе - False;
- **\_\_neg\_\_**: 
    - функция для покомпонентного умножения вектора на -1; перегрузка унарного оператора `-`;
    - должна возвращать новый объект типа Vector, компоненты которого - компоненты данного вектора, взятые с противоположным знаком;
- **\_\_add\_\_**: 
    - перегрузка бинарного оператора `+`;
    - покомпонентно складывает вектор с любым итерируемым объектом, если это не вызывает исключения;
    - если размеры векторов не совпадают, меньший вектор необходимо дополнить нулями до размера большего вектора;
    - если сложение вызывает исключение, необходимо вернуть значение NotImplemented;
    - возвращает новый вектор, компонентами которого являются суммы соответствующих компонент операндов;
- **\_\_radd\_\_**: 
    - перегрузка бинарного оператора `+`, при расположении объекта типа Vector справа;
    - обладает поведением функции \_\_add\_\_; Реализуйте её через функцию \_\_add\_\_;
- **\_\_sub\_\_**:
    - перегрузка бинарного оператора `-`;
    - обладает поведением функции \_\_add\_\_; Реализуйте её через функцию \_\_add\_\_;
- **\_\_mul\_\_**:
    - перегрузка бинарного оператора `*`;
    - покомпонентно умножает вектор на скаляр;
    - если тип переданного значения отличен от Real, необходимо вернуть значение NotImplemented;
    - возвращает новый вектор, компонентами которого являются произведения соответствующих компонент исходного вектора на скаляр;
- **\_\_rmul\_\_**:
    - перегрузка бинарного оператора `*`, при расположении объекта типа Vector справа;
    - обладает поведением функции \_\_mul\_\_; Реализуйте её через функцию \_\_mul\_\_;
- **\_\_matmul\_\_**:
    - перегрузка бинарного оператора `@`; реализует скалярное произведение вектора на любой итерируемый объект;
    - если размеры векторов не совпадают, меньший вектор необходимо дополнить нулями до размера большего вектора;
    - если скалярное умножение вызывает исключение, необходимо вернуть значение NotImplemented;
    - возвращает число - значение скалярного произведения; 
- **\_\_rmatmul\_\_**:
    - перегрузка бинарного оператора `@`; реализует скалярное произведение вектора на любой итерируемый объект;
    - обладает поведением функции \_\_matmul\_\_; Реализуйте её через функцию \_\_matmul\_\_;
- **\_\_truediv\_\_**:
    - перегрузка бинарного оператора `/`;
    - обладает поведением функции \_\_mul\_\_; Реализуйте её через функцию \_\_mul\_\_;
